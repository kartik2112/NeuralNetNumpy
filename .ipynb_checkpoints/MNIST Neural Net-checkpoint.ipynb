# Multi-Class Classification Neural Networks Training

The back propagation algorithm with regularisation has been implemented as explained in Andrew NG's Machine Learning Course. However, here we are using the original MNIST dataset created by Yann LeCun.

This one is flexible and works for any number of layers.

Highest Accuracy Achieved: 91.67%

```python
import numpy as np
import gzip
import idx2numpy
import matplotlib.pyplot as plt
```

First, we unip the files and extract the data

```python
with gzip.open('./data/train-images-idx3-ubyte.gz','r') as f:
    train_X = idx2numpy.convert_from_file(f).reshape(-1,784)
with gzip.open('./data/train-labels-idx1-ubyte.gz','r') as f:
    train_y = idx2numpy.convert_from_file(f)
with gzip.open('./data/t10k-images-idx3-ubyte.gz','r') as f:
    test_X = idx2numpy.convert_from_file(f).reshape(-1,784)
with gzip.open('./data/t10k-labels-idx1-ubyte.gz','r') as f:
    test_y = idx2numpy.convert_from_file(f)
```

```python
train_X.shape, train_y.shape, test_X.shape, test_y.shape
```

We define all the functions that would be needed for computation.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoidGrad(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

We use logistic regression for multi-class classification. This would work for any number of layers.

```python
def linearCostFunction(X,y,theta,layerSizes,wd):    
    thetas = []
    m = X.shape[0]
    hotEncY = np.zeros((m,10))
    hotEncY[np.arange(m),y.flatten()] = 1
    
    noOfLayers = len(layerSizes)
    layerStart = 0
    for i in range(noOfLayers-1):
        thetas.append(theta[layerStart:layerStart+(layerSizes[i]+1)*layerSizes[i+1]].reshape((layerSizes[i+1],layerSizes[i]+1)))
        layerStart += (layerSizes[i]+1)*layerSizes[i+1]
#     for i in range(noOfLayers-1):
#         print(thetas[i].shape)
    Zs, As = [], []
    Zs.append(0)
    As.append(X)
    
    # Forward Propagation
    for i in range(noOfLayers-1):
        lastA = As[-1]
        Z = np.append(np.ones((lastA.shape[0],1)),lastA,axis=1) @ np.transpose(thetas[i])
        Zs.append(Z)
        A = sigmoid(Z)
        As.append(A)
    
    # Cost Calculation
    regPart = 0
    for i in range(noOfLayers-1):
        regPart += (thetas[i][1:] ** 2).sum()
    cost = 1 / m * (- hotEncY * np.log(As[-1]) - (1-hotEncY) * np.log(1-As[-1])).sum() + wd / (2 * m) * regPart
    
    # Backward Propagation
    prevLayerDelta = np.append(np.zeros((m,1)),(As[-1] - hotEncY),axis=1)
    Deltas = []
    for i in range(noOfLayers-2, -1,-1):
        Delta = (np.transpose(prevLayerDelta[:,1:]) @ np.append(np.ones((m,1)),As[i],axis=1)) / m
        WD = np.append(np.zeros((1, thetas[i].shape[1])), wd / m * thetas[i][1:,:],axis=0)
        Delta += WD
        Deltas.append(Delta.flatten())
        if i != 0:
            prevLayerDelta = (prevLayerDelta[:,1:] @ thetas[i]) * sigmoidGrad(np.append(np.ones((m,1)),Zs[i],axis=1))
    Deltas1 = np.array([])
    for i in range(len(Deltas)-1,-1,-1):
        Deltas1 = np.append(Deltas1,Deltas[i])
    return (cost, Deltas1)
    
```

```python
def predictionAccuracy(X,y,theta,layerSizes):
    thetas = []
    m = X.shape[0]
    hotEncY = np.zeros((m,10))
    hotEncY[np.arange(m),y.flatten()] = 1
    
    noOfLayers = len(layerSizes)
    layerStart = 0
    for i in range(noOfLayers-1):
        thetas.append(theta[layerStart:layerStart+(layerSizes[i]+1)*layerSizes[i+1]].reshape((layerSizes[i+1],layerSizes[i]+1)))
        layerStart += (layerSizes[i]+1)*layerSizes[i+1]
        
        
    lastA = X
    
    
    
    # Forward Propagation
    for i in range(noOfLayers-1):
        lastA = sigmoid(np.append(np.ones((lastA.shape[0],1)),lastA,axis=1) @ np.transpose(thetas[i]))
#     print(lastA)
    return (lastA.argmax(axis=1) == y).sum() / y.shape[0]
```

```python
from numpy import random
def randomThetas(size):
    eps = 0.12
    return random.rand(size) * 2 * eps - eps
```

Here, we are constructing a 2-layer NN with 25 hidden units and 10 output units corresponding to 10 outputs

```python
layers = [784,25,10]
m = train_X.shape[0]

totSize = 0
for i in range(len(layers)-1):
    totSize += (layers[i]+1)*layers[i+1]
print('Parameters that will be trained: ',totSize)
```

We evaluate whether our implementation is correct by verifying the gradients with the numerical gradient calculations.

```python
def numericalGradients(X,y,thetas,layers):
    eps = 0.0001
    Deltas = np.zeros(thetas.shape)
    for p in range(len(thetas)):
        print(p,'\r',end='')
        thetas[p] += eps
        c1,_ = linearCostFunction(X,y,thetas,layers,0)
        thetas[p] -= 2*eps
        c2,_ = linearCostFunction(X,y,thetas,layers,0)
        thetas[p] += eps
        Deltas[p] = (c1-c2)/(2*eps)
    return Deltas
```

```python
from sklearn.metrics import mean_absolute_error
thetas = randomThetas(totSize)
(cost,deltas1) = linearCostFunction(train_X[1:100],train_y[1:100],thetas,layers,0)
```

```python
deltas2 = numericalGradients(train_X[1:100],train_y[1:100],thetas,layers)
```

```python
mean_absolute_error(deltas2,deltas1)
```

Now, we start training the data using various parameters and see which one gives us the best test accuracy.

```python
num_epochs = 200
lr = 0.1
thetas = randomThetas(totSize)
costs = []
for i in range(num_epochs):
    (cost,deltas) = linearCostFunction(train_X,train_y,thetas,layers,0.5)
    costs.append(cost)
    print("Epoch:",i, ", Cost: ",cost)
    thetas -= lr * deltas
predictionAccuracy(test_X,test_y,thetas,layers)
```

## SGD with Momentum and Learning Rate Annealing

This technique is referred from FastAI Lessons

```python
layers = [784,300,200,100,10]
m = train_X.shape[0]

totSize = 0
for i in range(len(layers)-1):
    totSize += (layers[i]+1)*layers[i+1]
print('Parameters that will be trained: ',totSize)
```

```python
num_epochs = 500
# lr = 0.1 to 0.00001
lr = [0.5 for x in range(100)]+[(0.5-0.001) * x / 400 + 0.001 for x in range(400,0,-1)]
thetas = randomThetas(totSize)
prevDeltas = np.zeros(thetas.shape)
costs = []
mom = 0.5
for i in range(num_epochs):
    (cost,deltas) = linearCostFunction(train_X,train_y,thetas,layers,1)
    costs.append(cost)
    thetas -= lr[i] * (mom * deltas + (1 - mom) * prevDeltas)
    print("Epoch:",i, ", Cost: ",cost, ", Accuracy: ",predictionAccuracy(test_X,test_y,thetas,layers))
    prevDeltas = deltas
```

```python
plt.plot(costs)
```
